```python
#Create a new column categorizing the time the crime occurred

def time_of_day(x):
    if x<400:
        return 'Late Night'
    elif x<800:
        return 'Early Morning'
    elif x<1200:
        return 'Late Morning'
    elif x<1600:
        return 'Afternoon'
    elif x<2000:
        return 'Evening'
    elif x<2400:
        return 'Night'
    
df['Time of Day'] = df['TIME OCC'].apply(time_of_day)
```

```python
#Compare the above amounts/percentages with a histogram

plt.figure(figsize= (15,8))
sns.histplot(data=df['TIME OCC'], binwidth = 100, color = 'green')
plt.title('Times When Crimes Occurred', fontsize=18)
plt.xlabel('Time', fontsize=15)
plt.ylabel('Number of Crimes', fontsize=15)

#The values don't match with the findings above
#E.g., Late Morning (8am-12pm) is here greater than Night (8pm-12am) 
```

```python
#Adjust the histogram by specifying binrange also

plt.figure(figsize= (15,8))
sns.histplot(data=df['TIME OCC'], binwidth = 100, binrange = (0, 2400), color = 'green')
plt.title('Times When Crimes Occurred', fontsize=18)
plt.xlabel('Time', fontsize=15)
plt.ylabel('Number of Crimes', fontsize=15)

#Results appear correct now, data is sorted into the right bins

#Most crimes occurred in the Evening category (4pm-8pm)
#But the time interval with the most crimes is 12pm-1pm 
#Lowest number of crimes occurred from 5am-6am
```
